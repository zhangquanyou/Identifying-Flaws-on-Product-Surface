```python
from matplotlib import pyplot as plt
from PIL import ImageFont, ImageDraw, Image
import cv2 as cv
import numpy as np
import os
import pickle
import time
```

```python
image = cv.imread(filename = 'Img/11.JPG',
                   flags = cv.IMREAD_GRAYSCALE)
```

```python
SIFT = cv.xfeatures2d.SIFT_create()
AKAZE = cv.AKAZE_create()
KAZE = cv.KAZE_create()
STAR = cv.xfeatures2d.StarDetector_create()#
BRIEF = cv.xfeatures2d.BriefDescriptorExtractor_create()
BRISK = cv.BRISK_create()
#FREAK = cv.xfeatures2d.FREAK_create()
SURF = cv.xfeatures2d.StarDetector_create()#

ORB = cv.ORB_create()
ORB.setScaleFactor(1.7)
ORB.setEdgeThreshold(63)

t0 = time.time()
SIFT_keypoints = SIFT.detect(image, None)
t1 = time.time()
print("SIFT：Number of keypoints Detected:", len(SIFT_keypoints), "\n")
print("SIFT：Time=:", t1-t0, "\n")

t0 = time.time()
AKAZE_keypoints = AKAZE.detect(image, None)
t1 = time.time()
print("AKAZE：Number of keypoints Detected:", len(AKAZE_keypoints), "\n")
print("AKAZE：Time=:", t1-t0, "\n")

t0 = time.time()
KAZE_keypoints = KAZE.detect(image, None)
t1 = time.time()
print("KAZE:Number of keypoints Detected:", len(KAZE_keypoints), "\n")
print("KAZE：Time=:", t1-t0, "\n")

t0 = time.time()
STAR_keypoints = STAR.detect(image, None)
t1 = time.time()
#BRIEF_keypoints=BRIEF.detect(image,None)
print("STAR:Number of keypoints Detected:", len(STAR_keypoints), "\n")
print("STAR：Time=:", t1-t0, "\n")

t0 = time.time()
BRISK_keypoints = BRISK.detect(image, None)
t1 = time.time()
print("BRISK:Number of keypoints Detected:", len(BRISK_keypoints), "\n")

#t0 = time.time()
#FREAK_keypoints = FREAK.detect(image, None)
#t1 = time.time()
#print("FREAK:Number of keypoints Detected:", len(FREAK_keypoints), "\n")
#print("FREAK：Time=:", t1-t0, "\n")

t0 = time.time()
ORB_keypoints = ORB.detect(image, None)
t1 = time.time()
print("ORB:Number of keypoints Detected:", len(ORB_keypoints), "\n")
print("ORB：Time=:", t1-t0, "\n")

t0 = time.time()
SURF_keypoints = SURF.detect(image, None)
t1 = time.time()
print("SURF:Number of keypoints Detected:", len(SURF_keypoints), "\n")
print("SURF：Time=:", t1-t0, "\n")
```

```python
index = []

for point in SIFT_keypoints:
    temp = (point.pt,
            point.size,
            point.angle,
            point.response,
            point.octave, 
            point.class_id)
    
    index.append(temp)
    
# File name
filename = "Outputs/SIFT-keypoints.txt"

#Delete a file if it exists
if os.path.exists(filename):
    os.remove(filename)

# Open a file
file = open(filename, "wb")

# Write 
file.write(pickle.dumps(index))

# Close a file
file.close()
keypoints, descriptors = SIFT.compute(image, SIFT_keypoints)
image1 = cv.drawKeypoints(image = image,
                         keypoints = keypoints[:500],
                         outImage = None,
                         flags = cv.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS
                         )
```
